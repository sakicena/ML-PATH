```python
# choice() in random in numpy
import numpy as np
x=np.random.choice([4,1,2,5],p=[0.25,0.25,0.05,0.45],size=(100))
x
```

```python
# shuffle and permutation
import numpy as np
x=np.array([7,10,3,1,5])
np.random.permutation(x)
print(x)
np.random.shuffle(x)
print(x)
```

```python
# seaborn module
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot([5,2,3,4],hist=False)
plt.show()
```

```python
# normal distribution
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
a=np.random.normal(loc=1,size=(2,2),scale=2) #loc is mean and scale is standard deviation
print(a)
sbs.distplot(np.random.normal(loc=1,scale=2,size=(1000)),hist=False)
plt.show()
```

```python
# binomial distribution Discrete Distribution.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
a=np.random.binomial(n=10,p=0.5,size=10)
print(a)
sbs.distplot(np.random.binomial(n=100,p=0.5,size=1000),hist=False,label='Binomial dst')
sbs.distplot(np.random.normal(loc=50,scale=5,size=1000),hist=False,label='Normal dst')
plt.title('Normal vs Binomial Curve')
plt.show()
sbs.distplot(np.random.binomial(n=100,p=0.5,size=1000),hist=True,label='Binomial dst',kde=False)
sbs.distplot(np.random.normal(loc=50,scale=5,size=1000),hist=True,label='Normal dst',kde=False)
plt.title('Normal vs Binomial Hist')
plt.show()
```

```python
# Poisson Distribution is a Discrete Distribution.
# It estimates how many times an event can happen in a specified time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
a=np.random.poisson(lam=2,size=10)
print(a)
sbs.distplot(np.random.poisson(lam=2,size=1000),kde=False,label='Poisson Dist')
plt.title('Poisson Dist')
plt.show()
sbs.distplot(np.random.poisson(lam=50,size=1000),hist=False,label='Poisson Dist')
sbs.distplot(np.random.normal(loc=50,scale=7,size=1000),hist=False,label='Normal Dist')
plt.title('Poisson Dist vs Normal Dist')
plt.show()
sbs.distplot(np.random.poisson(lam=10,size=1000),hist=False,label='Poisson Dist')
sbs.distplot(np.random.binomial(n=1000,p=0.01,size=1000),hist=False,label='Binomial Dist')
plt.title('Poisson Dist vs Binomial Dist')
plt.show()
```

```python
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot(np.random.uniform(size=1000),hist=False)
plt.title('Uniform Distribution')
plt.show()
```

```python
#logistic Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot(np.random.logistic(size=1000),hist=False)
plt.title('Logistic Distribution')
plt.show()
sbs.distplot(np.random.logistic(size=1000),hist=False,label='Logistic Dist')
sbs.distplot(np.random.normal(scale=2,size=1000),hist=False,label='Normal Dist')
plt.title('Logistic Dist vs Normal Dist')
plt.show()
```

```python
# Multinomial Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot(np.random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6],size=1),hist=False)
plt.title('Multinomial')
plt.show()
```

```python
# Exponential Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot(np.random.exponential(size=1000),hist=False)
plt.title('Exponential Distribution')
plt.show()
```

```python
#Chi Square Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot(np.random.chisquare(df=2,size=1000),hist=False)
plt.title('Chi Square Distribution')
plt.show()
```

```python
# Rayleigh Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.rayleigh(size=1000), hist=False)
plt.title('Rayleigh Distribution')
plt.show()
```

```python
# Pareto Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
sbs.distplot(random.pareto(a=5, size=50), kde=False)
plt.title('Pareto Distribution')
plt.show()
```

```python
# Zipf Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)
plt.title('Zipf Distribution')
plt.show()
```
