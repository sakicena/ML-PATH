```python
#drop na cells to another df
import pandas as pd
df=pd.read_csv('dirtydata.csv')
df=df.dropna()
df

#drop na with same df
df.dropna(inplace=True)
df
```

```python
#fill na cells for all columns
import pandas as pd
df=pd.read_csv('dirtydata.csv')
df.fillna(130,inplace=True)
df

#fill na cells for selected columns
df1=pd.read_csv('dirtydata.csv')
df1['Calories'].fillna(130,inplace=True)
df1

#fill na cells with median
df2=pd.read_csv('dirtydata.csv')
x=df2['Calories'].median()
df2['Calories'].fillna(x,inplace=True)
df2

#fill na cells with mode
df3=pd.read_csv('dirtydata.csv')
x=df3['Calories'].mode()[0]
df3['Calories'].fillna(x,inplace=True)
df3

#fill na cells with mean
df4=pd.read_csv('dirtydata.csv')
x=df4['Calories'].mean()
df4['Calories'].fillna(x,inplace=True)
df4
```

```python
#date wrong format in csv
import pandas as pd
df=pd.read_csv('dirtydata.csv')
df['Date']=pd.to_datetime(df['Date'])
df

#removing NaT from df
df.dropna(subset=['Date'],inplace=True)
df
```

```python
#replacing index
import pandas as pd
df=pd.read_csv('dirtydata.csv')
for x in df.index:
    if df.loc[x,'Duration']>120:
        df.loc[x,'Duration']=120
df


#drop wrong rows
import pandas as pd
df1=pd.read_csv('dirtydata.csv')
for x in df1.index:
    if df1.loc[x,'Duration']>120:
        df1.drop(x,inplace=True)
df1

```

```python
# drop duplicates
import pandas as pd
df=pd.read_csv('dirtydata.csv')
df.drop_duplicates(inplace=True)
df
```

```python
# Cleaning Data
import pandas as pd
df=pd.read_csv('dirtydata.csv')
x=df['Calories'].mean()
df['Calories'].fillna(x,inplace=True)
df['Date']=pd.to_datetime(df['Date'])
df.dropna(subset=['Date'],inplace=True)
df.drop_duplicates(inplace=True)
for x in df.index:
    if df.loc[x,'Duration']>120:
        df.drop(x,inplace=True)
df
```
