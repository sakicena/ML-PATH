```python
# Plotting
import numpy as np
import matplotlib.pyplot as plt
x=np.array([15,6])
y=np.array([1,10])
plt.plot(x,y,'o')
plt.title('Dots')
plt.show()
plt.plot(x,y)
plt.title('Line')
plt.show()
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.title('Points Dots')
plt.show()

plt.plot(ypoints)
plt.title('Single Axis Dots')
plt.show()
```

```python
# Markers
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o')
# example of markers : circle=o , point=. ,pixel = ,
#marker attribute  for type of dots 
plt.title('Marker Attribute')
plt.show()
plt.plot(ypoints, 'o--') # line attributes '-',':','--','-.'
plt.title('Marker Attribute with line string attribute')  
plt.show()
plt.plot(ypoints, 'o-.k') #colors ex: red=r,blue=b,green=g,cyan=c,magenta=m,yellow=y,black=k,white=w
plt.title('Marker Attribute with color and line attributes')
plt.show()
```

```python
# Marker size and colors
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,  'o-.y', ms = 20,linewidth=5) # ms as marker size with linewidth
plt.title('Marker Attribute with marker size ms and line width')
plt.show()
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r') #mec as marker edge color
plt.title('Marker Attribute with marker edge color')
plt.show()
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r') #mfc as marker edge inside color
plt.title('Marker Attribute with marker edge inside  color')
plt.show()
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'b', mfc = 'y')#mfc mec ms attributes
plt.title('Marker Attribute with marker edge inside  color,marker edge and marker size')
plt.show()
```

```python
# Create labels and title and fontdict
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font = {'family':'serif','color':'blue','size':15}
font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y)
plt.title("Sports Watch Data",fontdict=font,loc='center')
plt.xlabel("Average Pulse",fontdict=font1)
plt.ylabel("Calorie Burnage",fontdict=font2)

plt.show()
```

```python
# Adding Grid and some properties
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font = {'family':'serif','color':'blue','size':15}
font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y)
plt.title("Sports Watch Data",fontdict=font,loc='center')
plt.xlabel("Average Pulse",fontdict=font1)
plt.ylabel("Calorie Burnage",fontdict=font2)
plt.grid(linewidth = 1,linestyle='-.',color='g')
plt.show()
```

```python
# Scatter Plots
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
# Bar Chats
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y,height = 0.5)
plt.show()
plt.bar(x, y,width = 0.5)
plt.show()
```

```python
# Pie Charts
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"] #Labels
myexplode = [0.1, 0.05, 0.05, 0.05] #Explode
mycolors = ["black", "hotpink", "b", "#4CAF50"] #colors
mylabels = ["Apples", "Bananas", "Cherries", "Dates"] #Legend
plt.pie(y,labels=mylabels,startangle=0,explode=myexplode,shadow=True,colors=mycolors)
plt.legend(title = "Four Fruits:", loc='center') #plot legend with title
plt.show() 
```
